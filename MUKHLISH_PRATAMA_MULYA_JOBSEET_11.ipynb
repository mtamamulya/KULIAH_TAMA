```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}"
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Membuat data 1 sampai 100.000
data = list(range(1, 100001))
target = 99999

# Mengukur waktu Sequential Search
start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

# Mengukur waktu Binary Search
start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

```python
import random

def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Target {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Target {target} tidak ditemukan dalam data"

# 1. Generate 15 angka acak
data_acak = [random.randint(1, 100) for _ in range(15)]
print(f"Data acak: {data_acak}")

# 2. Urutkan data (Binary Search wajib menggunakan data yang terurut)
data_terurut = insertion_sort(data_acak.copy())
print(f"Data setelah diurutkan: {data_terurut}")

# 3. Input target dari user
try:
    target = int(input("\nMasukkan angka yang ingin dicari: "))

    # 4. Jalankan pencarian
    hasil = binary_search(data_terurut, target)
    print(hasil)
except ValueError:
    print("Mohon masukkan angka yang valid (integer).")
```

```python
def sequential_search_counted(data, target):
    comparisons = 0
    for i in range(len(data)):
        comparisons += 1 # Menghitung setiap perbandingan
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}. Jumlah perbandingan: {comparisons}"
    return f"Data {target} tidak ditemukan. Jumlah perbandingan: {comparisons}"

# Contoh Penggunaan
angka = [10, 20, 30, 40, 50]
print(sequential_search_counted(angka, 30))
```

```python
def binary_search_counted(data, target):
    left, right = 0, len(data) - 1
    iterations = 0

    while left <= right:
        iterations += 1 # Menghitung setiap langkah iterasi
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}. Jumlah iterasi: {iterations}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Data {target} tidak ditemukan. Jumlah iterasi: {iterations}"

# Contoh Penggunaan (Data harus terurut [cite: 86])
angka_urut = [10, 20, 30, 40, 50]
print(binary_search_counted(angka_urut, 40))
```

```python
def search_student_name():
    # Input daftar nama dipisahkan dengan koma
    input_nama = input("Masukkan daftar nama mahasiswa (pisahkan dengan koma): ")
    daftar_nama = [nama.strip() for nama in input_nama.split(",")]

    target = input("Masukkan nama yang ingin dicari: ")

    # Sequential search dengan case-insensitive [cite: 283, 495]
    found = False
    for i in range(len(daftar_nama)):
        if daftar_nama[i].lower() == target.lower():
            print(f"Nama '{target}' ditemukan di indeks {i}.")
            found = True
            break

    if not found:
        print(f"Nama '{target}' tidak ditemukan dalam daftar.")

# Jalankan fungsi
search_student_name()
```

```python
def search_price():
    barang = [("Buku", 5000), ("Pensil", 2000), ("Penggaris", 3000), ("Tas", 50000)]

    # 1. Urutkan berdasarkan harga (syarat Binary Search [cite: 170])
    barang.sort(key=lambda x: x[1])
    print("Daftar barang terurut harga:", barang)

    target_harga = int(input("Masukkan harga yang dicari: "))

    # 2. Binary Search pada harga
    left, right = 0, len(barang) - 1
    found = False

    while left <= right:
        mid = (left + right) // 2
        if barang[mid][1] == target_harga:
            print(f"Barang dengan harga {target_harga} ditemukan: {barang[mid][0]}")
            found = True
            break
        elif barang[mid][1] < target_harga:
            left = mid + 1
        else:
            right = mid - 1

    if not found:
        print("Barang dengan harga tersebut tidak ditemukan.")

search_price()
```

```python
import time

def compare_complexity():
    sizes = [100, 1000, 10000]
    print(f"{'Ukuran Data':<15} | {'Sequential (detik)':<20} | {'Binary (detik)':<20}")
    print("-" * 60)

    for size in sizes:
        data = list(range(size))
        target = size - 1 # Mencari elemen terakhir (Worst Case [cite: 82])

        # Waktu Sequential Search [cite: 381]
        start_seq = time.time()
        for i in range(len(data)):
            if data[i] == target: break
        end_seq = time.time()

        # Waktu Binary Search [cite: 384]
        start_bin = time.time()
        l, r = 0, len(data)-1
        while l <= r:
            m = (l+r)//2
            if data[m] == target: break
            elif data[m] < target: l = m + 1
            else: r = m - 1
        end_bin = time.time()

        print(f"{size:<15} | {end_seq - start_seq:<20.8f} | {end_bin - start_bin:<20.8f}")

compare_complexity()
```

```python
def mini_project_nilai():
    # Input data
    n = int(input("Berapa jumlah mahasiswa? "))
    data_mahasiswa = []

    for i in range(n):
        nama = input(f"Nama mahasiswa ke-{i+1}: ")
        nilai = int(input(f"Nilai {nama}: "))
        data_mahasiswa.append((nama, nilai))

    # Urutkan berdasarkan nilai untuk Binary Search [cite: 502, 509]
    data_mahasiswa.sort(key=lambda x: x[1])
    print("\nData Terurut (Nilai):", data_mahasiswa)

    target_nilai = int(input("\nMasukkan nilai yang ingin dicari: "))

    # Binary Search
    l, r = 0, len(data_mahasiswa) - 1
    found_idx = -1

    while l <= r:
        mid = (l + r) // 2
        if data_mahasiswa[mid][1] == target_nilai:
            found_idx = mid
            break
        elif data_mahasiswa[mid][1] < target_nilai:
            l = mid + 1
        else:
            r = mid - 1

    if found_idx != -1:
        print(f"Nilai {target_nilai} ditemukan! Milik mahasiswa: {data_mahasiswa[found_idx][0]}")
    else:
        print("Nilai tidak ditemukan dalam daftar.")

mini_project_nilai()
```
