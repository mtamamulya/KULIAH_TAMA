```python
%%writefile bubble_sort.py
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                # Tukar posisi
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

```python
%%writefile selection_sort.py
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        # Tukar data[i] dengan data[min_index]
        data[i], data[min_index] = data[min_index], data[i]
    return data

# Pengujian
angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

```python
%%writefile insertion_sort.py
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

# Pengujian
angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            # Bandingkan nilai (indeks ke-1 dari tuple)
            if data[j][1] < data[j + 1][1]:
                # Tukar posisi
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Pengujian
mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                # Untuk descending (besar ke kecil)
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                # Untuk ascending (kecil ke besar)
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Pengujian
barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]

# Urut berdasarkan nama (indeks 0) secara ascending
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))

# Urut berdasarkan harga (indeks 1) secara descending (reverse=True)
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

```python
def bubble_sort_descending(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            # Ubah tanda '>' menjadi '<' untuk urutan menurun
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Pengujian
angka = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort Descending:", bubble_sort_descending(angka.copy()))
```

```python
def selection_sort_max_to_end(data):
    n = len(data)
    for i in range(n):
        # Asumsikan elemen pertama (index 0) adalah max
        max_index = 0

        # Cari elemen terbesar di array yang belum terurut (0 sampai n-1-i)
        last_index = n - 1 - i
        for j in range(1, last_index + 1):
            if data[j] > data[max_index]:
                max_index = j

        # Tukar elemen terbesar dengan elemen di posisi akhir yang belum terurut
        data[last_index], data[max_index] = data[max_index], data[last_index]
    return data

# Pengujian
angka = [64, 25, 12, 22, 11]
print("Selection Sort (Max ke Akhir):", selection_sort_max_to_end(angka.copy()))
```

```python
def insertion_sort_string(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        # Python otomatis membandingkan string secara leksikografis (abjad)
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

# Pengujian
nama = ["Zainab", "Andi", "Budi", "Eka", "Dian"]
print("Insertion Sort (String):", insertion_sort_string(nama.copy()))
```

```python
%%writefile nilai.py
def urutkan_nilai_mahasiswa(data):
    # Mengurutkan list of tuples [(Nama, Nilai)] berdasarkan Nilai (descending)
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        # Bandingkan berdasarkan nilai (indeks 1), urut desc (key > data[j])
        while j >= 0 and data[j][1] < key[1]:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data
```

```python
import random

# Modifikasi fungsi untuk mengembalikan jumlah perbandingan (count)
def bubble_count(data):
    count = 0
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return count

def selection_count(data):
    count = 0
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            count += 1
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return count

def insertion_count(data):
    count = 0
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            count += 1 # Hitung perbandingan
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return count

# Eksekusi Analisis
sizes = [10, 50, 100]
print(f"{'Ukuran':<10} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 50)

for size in sizes:
    data_awal = [random.randint(1, 1000) for _ in range(size)]

    c_bubble = bubble_count(data_awal.copy())
    c_select = selection_count(data_awal.copy())
    c_insert = insertion_count(data_awal.copy())

    print(f"{size:<10} | {c_bubble:<10} | {c_select:<10} | {c_insert:<10}")
```

```python
def sort_barang(data, key_idx, is_desc):
    # Menggunakan Bubble Sort yang dimodifikasi
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            condition = False
            if is_desc:
                if data[j][key_idx] < data[j + 1][key_idx]: condition = True
            else:
                if data[j][key_idx] > data[j + 1][key_idx]: condition = True

            if condition:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

def main_proyek():
    daftar_barang = []

    # Input Data
    print("=== Masukkan Data Barang (Ketik 'stop' di nama untuk selesai) ===")
    while True:
        nama = input("Nama Barang: ")
        if nama.lower() == 'stop':
            break
        try:
            harga = int(input("Harga Barang: "))
            daftar_barang.append((nama, harga))
        except ValueError:
            print("Harga harus angka!")

    if not daftar_barang:
        print("Tidak ada data.")
        return

    # Pilihan Sorting
    print("\nOpsi Pengurutan:")
    print("1. Berdasarkan Nama (A-Z)")
    print("2. Berdasarkan Nama (Z-A)")
    print("3. Berdasarkan Harga (Termurah-Termahal)")
    print("4. Berdasarkan Harga (Termahal-Termurah)")

    pilihan = input("Pilih nomor (1-4): ")

    if pilihan == '1':
        hasil = sort_barang(daftar_barang, 0, False)
    elif pilihan == '2':
        hasil = sort_barang(daftar_barang, 0, True)
    elif pilihan == '3':
        hasil = sort_barang(daftar_barang, 1, False)
    elif pilihan == '4':
        hasil = sort_barang(daftar_barang, 1, True)
    else:
        print("Pilihan tidak valid")
        return

    print("\n=== Hasil Pengurutan ===")
    for b in hasil:
        print(f"{b[0]}: Rp {b[1]}")

# Jalankan program jika file dieksekusi langsung
if __name__ == "__main__":
    main_proyek()
```
