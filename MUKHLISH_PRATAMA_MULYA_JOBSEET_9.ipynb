```python
import math

def hitung_luas_lingkaran(jari_jari):
    luas = math.pi * (jari_jari ** 2)
    return luas

# Uji dengan jari-jari 7 dan 10
r1 = 7
r2 = 10

print(f"Luas lingkaran dengan jari-jari {r1} adalah: {hitung_luas_lingkaran(r1):.2f}")
print(f"Luas lingkaran dengan jari-jari {r2} adalah: {hitung_luas_lingkaran(r2):.2f}")
```

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi

# Uji Celcius ke Fahrenheit
suhu_c = 30
hasil_f = konversi.celcius_ke_fahrenheit(suhu_c)
print(f"{suhu_c} Celcius = {hasil_f} Fahrenheit")

# Uji Fahrenheit ke Celcius
suhu_f = 86
hasil_c = konversi.fahrenheit_ke_celcius(suhu_f)
print(f"{suhu_f} Fahrenheit = {hasil_c} Celcius")
```

```python
%%writefile laporan.py

def tampilkan_skor(nama, skor_akhir):
    print("\n" + "="*30)
    print("LAPORAN HASIL KUIS")
    print("="*30)
    print(f"Peserta : {nama}")
    print(f"Skor Akhir : {skor_akhir}")

    if skor_akhir >= 80:
        print("Predikat : Sangat Baik!")
    elif skor_akhir >= 60:
        print("Predikat : Cukup Baik")
    else:
        print("Predikat : Belajar Lagi Ya!")
```

```python
import random
import laporan

def main():
    print("=== KUIS MATEMATIKA SEDERHANA ===")
    nama = input("Masukkan Nama Kamu: ")
    skor = 0
    jumlah_soal = 5

    for i in range(jumlah_soal):
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        jawaban_benar = a + b

        try:
            jawab = int(input(f"Soal {i+1}: Berapa {a} + {b}? "))
            if jawab == jawaban_benar:
                print("Benar!")
                skor += 20  # 20 poin per soal (total 100)
            else:
                print(f"Salah. Jawabannya {jawaban_benar}")
        except ValueError:
            print("Input tidak valid (harus angka). Dianggap salah.")

    # Memanggil modul laporan.py
    laporan.tampilkan_skor(nama, skor)

if __name__ == "__main__":
    main()
```
