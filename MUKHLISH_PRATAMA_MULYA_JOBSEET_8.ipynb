```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

```python
x = 10

def ubah():
    x = 5
    print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

```python
kali = lambda x, y: x * y
print("Hasil perkalian:", kali(4, 5))
```

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

```python
def hitung_rata(data):
    return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("Rata-rata nilai:", rata)
```

```python
print("--- 1. Fungsi Sederhana ---")
def perkenalan(nama, prodi):
    print(f"Halo, nama saya {nama}, saya dari {prodi}.")

perkenalan("Andi", "Teknologi Rekayasa Komputer")
perkenalan("Budi", "Teknik Informatika")
perkenalan("Citra", "Teknik Elektro")
print()
```

**TUGAS 2**

```python
print("--- 2. Fungsi Parameter & Return ---")
def luas_persegi(panjang, lebar):
    return panjang * lebar

print(f"Luas (10, 5): {luas_persegi(10, 5)}")
print(f"Luas (7, 3): {luas_persegi(7, 3)}")
print(f"Luas (12, 12): {luas_persegi(12, 12)}")
print()
```

**TUGAS 3**

```python
print("--- 3. Default Argument ---")
def salam(waktu="pagi", nama="Mahasiswa"):
    print(f"Selamat {waktu}, {nama}!")

salam("sore", "Dosen")
salam(waktu="malam")
salam()
print()
```

**TUGAS 4**

```python
print("--- 4. Fungsi *args ---")
def jumlahkan_semua(*angka):
    return sum(angka)

print("Total (3 bil):", jumlahkan_semua(10, 20, 30))
print("Total (5 bil):", jumlahkan_semua(1, 2, 3, 4, 5))
print("Total (7 bil):", jumlahkan_semua(100, 200, 300, 400, 500, 600, 700))
print()
```

**TUGAS 5**

```python
print("--- 5. Fungsi **kwargs ---")
def info_barang(**data):
    for key, value in data.items():
        print(f"{key}: {value}")

info_barang(Nama="Buku", Harga=15000, Stok=30)
print("-" * 20)
info_barang(Nama="Laptop", Processor="i7", RAM="16GB", Harga=15000000)
print()
```

**TUGAS 6**

```python
print("--- 6. Scope Variabel ---")
x = 100

def ubah():
    x = 50
    print(f"Nilai di dalam fungsi (lokal): {x}")

ubah()
print(f"Nilai di luar fungsi (global): {x}")
print()
```

**TUGAS 7**

```python
print("--- 7. Fungsi Lambda ---")
kuadrat = lambda x: x ** 2

keliling_lingkaran = lambda r: 2 * 3.14 * r

print(f"Kuadrat dari 5: {kuadrat(5)}")
print(f"Keliling lingkaran (r=7): {keliling_lingkaran(7)}")
print()
```

**TUGAS 8**

```python
print("--- 8. Fungsi Decorator ---")
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, ini fungsi utama!")

halo()
print()
```

**TUGAS 9**

```python
print("--- 9. Studi Kasus Program Nilai ---")

def input_nilai():
    return [80, 75, 90, 65, 85]

def hitung_rata(data):
    return sum(data) / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

hitung_nilai_akhir = lambda tugas, ujian: (tugas * 0.4) + (ujian * 0.6)

daftar_nilai = input_nilai()
rata_rata = hitung_rata(daftar_nilai)
status = status_kelulusan(rata_rata)
nilai_akhir_mahasiswa = hitung_nilai_akhir(80, 90)

print(f"Daftar Nilai: {daftar_nilai}")
print(f"Rata-rata: {rata_rata}")
print(f"Status: {status}")
print(f"Nilai Akhir Bobot (Tugas 80, Ujian 90): {nilai_akhir_mahasiswa}")
```
